```python
import pandas as pd

file_path = "/content/drive/My Drive/MaternalHealthModel/maternal_dataset_csv.csv"
df = pd.read_csv(file_path)

# Preview the dataset
print("Shape:", df.shape)
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of age
plt.figure(figsize=(6,4))
sns.histplot(df['age'], kde=True)
plt.title('Age Distribution')
plt.show()

# Pairplot of selected features
# Removed 'Gravidity', 'Blood Pressure', 'Hemoglobin', 'Fetal Heart Rate'
# as they were not found in the dataframe.
# sns.pairplot(df[['Age', 'Gravidity', 'Blood Pressure', 'Hemoglobin', 'Fetal Heart Rate']])
# plt.show()
```

```python
# Show top numeric columns with missing values count
numeric_df = df.select_dtypes(include=['float64', 'int64'])
missing_counts = numeric_df.isnull().sum()
missing_counts[missing_counts < len(df) * 0.5]  # keep columns with <50% missing
```

```python
# Select only the relevant columns
selected_cols = [
    'age',
    'no_pregnancy',
    'duration_of_pregnancy_weeks_',
    'number_of_prior_pregnancies_with_live_births',
    'number_of_prior_pregnancies_with_stillbirths',
    'infant_weight_of_4kg_and_above',
    'height_cm',
    'weight_kg_v1',
    'bp_is_greater_than_140_90_v1',
    'newborn_birth_weight_kg_',
    'total_antenatal_visits'
]

df_selected = df[selected_cols]

# Fill any remaining missing values with column mean
df_selected = df_selected.fillna(df_selected.mean())

# Normalize the features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_selected)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Reduce to 2 dimensions for plotting
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

plt.figure(figsize=(8,6))
plt.scatter(pca_result[:,0], pca_result[:,1], alpha=0.5)
plt.title("PCA Projection of Maternal Health Data")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

```python
from sklearn.cluster import KMeans

# Try clustering into 3 groups (you can test more later)
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(scaled_data)

# Plot with clusters
plt.figure(figsize=(8,6))
plt.scatter(pca_result[:,0], pca_result[:,1], c=clusters, cmap='viridis', alpha=0.6)
plt.title("K-Means Clustering of Maternal Risk Profiles")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.colorbar(label='Cluster')
plt.show()
```

```python
# Assign clusters to original cleaned DataFrame
df_selected['Cluster'] = clusters

# Get mean values of each feature per cluster
cluster_profile = df_selected.groupby('Cluster').mean()

# Display the profile
cluster_profile.T  # Transpose for better readability
```

```python
import matplotlib.pyplot as plt

cluster_profile.T.plot(kind='bar', figsize=(12,6))
plt.title('Feature Averages per Cluster (Maternal Risk Groups)')
plt.ylabel('Average Value')
plt.xlabel('Feature')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Cluster', labels=['High Risk (0)', 'Low Risk (1)', 'Moderate Risk (2)'])
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Normalize the profile for radar chart
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
radar_data = scaler.fit_transform(cluster_profile)

# Setup
features = cluster_profile.columns.tolist()
labels = ['High Risk (0)', 'Low Risk (1)', 'Moderate Risk (2)']
angles = np.linspace(0, 2 * np.pi, len(features), endpoint=False).tolist()
angles += angles[:1]  # Complete the loop

# Plot
fig, ax = plt.subplots(figsize=(8,8), subplot_kw=dict(polar=True))
for i in range(3):
    values = radar_data[i].tolist()
    values += values[:1]  # Complete the loop
    ax.plot(angles, values, label=labels[i])
    ax.fill(angles, values, alpha=0.1)

ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_thetagrids(np.degrees(angles[:-1]), features, fontsize=9)
plt.legend(loc='upper right')
plt.title("Risk Cluster Profiles (Radar Chart)")
plt.show()
```

```python
import seaborn as sns

sns.countplot(x=clusters, palette='viridis')
plt.title("Number of Mothers per Cluster")
plt.xlabel("Cluster")
plt.ylabel("Count")
plt.xticks([0, 1, 2], ['High Risk (0)', 'Low Risk (1)', 'Moderate Risk (2)'])
plt.show()
```
